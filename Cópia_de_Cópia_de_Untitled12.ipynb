```python
lst_df_248_dias = []
```

```python
for item in np.split(df_248_dias,248):
  lst_df_248_dias.append(item.values)
```

```python
lst_df_248_dias[0]
```

```python
y_fech = []
y_max = []
y_min = []
lst_open_todos = []
max_open = []
dia = []
hora = []
lucro = []
max_open_todos = []
```

```python
for i in range(0,len(lst_df_248_dias)):
  for j in range(0,1):
    for k in range(1,36):
      lst_open_todos.append(lst_df_248_dias[i][k][2] -lst_df_248_dias[i][j][2])
      y_fech.append(lst_df_248_dias[i][k][5] -lst_df_248_dias[i][j][5])
      y_max.append(lst_df_248_dias[i][k][3] -lst_df_248_dias[i][j][3])
      y_min.append(lst_df_248_dias[i][k][4] -lst_df_248_dias[i][j][4])     
      dia.append(lst_df_248_dias[i][k][0])
      hora.append(lst_df_248_dias[i][k][1])
      lucro.append(lst_df_248_dias[i][k][5] - lst_df_248_dias[i][k][2])
      max_open.append(lst_df_248_dias[i][j][3] - lst_df_248_dias[i][j][2])
      max_open_todos.append(lst_df_248_dias[i][k][3] - lst_df_248_dias[i][k][2])
```

```python
df_regressão_1 = pd.DataFrame({'open': lst_open_todos,'max-open': max_open, 'max': y_max, 'min': y_min, 'fech': y_fech, 'lucro': lucro,'max_open_todos': max_open_todos, 'dia': dia, 'hora': hora})
df_regressão_1.head(36)
```

```python
from sklearn.model_selection import train_test_split
x1 = df_regressão_1.iloc[:, 0].values
y1 = df_regressão_1.iloc[:,2].values
x_train, x_teste, y_train, y_teste = train_test_split(x1,y1, test_size=0.25, random_state= 6)
```

```python
x1.shape, y1.shape
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
obj_linearregression = LinearRegression()
obj_linearregression.fit(x_train.reshape(-1,1), y_train.reshape(-1,1))
previsão_linear = obj_linearregression.predict(x_teste.reshape(-1,1))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_teste, previsão_linear), mean_squared_error(y_teste, previsão_linear), np.sqrt(mean_squared_error(y_teste, previsão_linear))
```

```python
from yellowbrick.regressor import residuals_plot
viz = residuals_plot(obj_linearregression,x_train.reshape(-1,1), y_train.reshape(-1,1), x_teste.reshape(-1,1), y_teste.reshape(-1,1))
```

Validação Cruzada

```python
resultados_regressão = []
resultados_regressão_1 = []
from sklearn.model_selection import cross_val_score, KFold
```

```python
for i in range(30):
  obj_kfold = KFold(n_splits=10 , shuffle= True, random_state = i)
  regressão = LinearRegression()
  scores = cross_val_score(regressão, x1.reshape(-1,1), y1.reshape(-1,1), cv=obj_kfold, scoring='neg_mean_squared_error')
  scores_1 = cross_val_score(regressão, x1.reshape(-1,1), y1.reshape(-1,1), cv=obj_kfold, scoring='r2')
  resultados_regressão.append(-scores.mean())
  resultados_regressão_1.append(scores_1.mean())
```

```python
np.sqrt(np.mean(resultados_regressão)), np.mean(resultados_regressão_1)
```
